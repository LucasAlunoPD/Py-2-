Notebook gerado automaticamente para entrega. Dados simulados, mas estrutura igual ao original.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# Dados simulados já carregados abaixo (wc: concat de masculino e feminino)
wc = pd.read_json('wc_simulado.json')

# mostrar as primeiras linhas
wc.head()
```

```
# 1) Quantas linhas e colunas tem o dataset?
wc_shape = wc.shape
wc_shape
```

```
# 2) Exiba estatísticas descritivas dos gols
wc[['gols_1','gols_2']].describe()
```

```
# 3) Quantas partidas por Copa (Masculina x Feminina)?
wc['copa'].value_counts()
```

```
# 4) Times que marcaram mais gols (somando gols_1 e gols_2 para cada time)
# transformar em formato long para contar gols por time
long = pd.concat([
    wc[['time_1','gols_1']].rename(columns={'time_1':'time','gols_1':'gols'}),
    wc[['time_2','gols_2']].rename(columns={'time_2':'time','gols_2':'gols'})
])
por_time = long.groupby('time')['gols'].sum().sort_values(ascending=False)
por_time.head(10)
```

```
# 5) Média de comparecimento por ano
media_comparecimento = wc.groupby('ano')['comparecimento'].mean().sort_index()
media_comparecimento.head(10)
```

```
# 6) Criar tabela pivô: média de gols por copa e rodada
pivot_gols = wc.pivot_table(index='rodada', columns='copa', values=['gols_1','gols_2'], aggfunc='mean')
pivot_gols
```

```
# 7) Quantas vitórias por time (considerando time_1 e time_2)
# função pra calcular resultado por linha

def resultado_linha(row):
    if row['gols_1'] > row['gols_2']:
        return row['time_1']
    elif row['gols_2'] > row['gols_1']:
        return row['time_2']
    else:
        return 'Empate'

wc['vencedor'] = wc.apply(resultado_linha, axis=1)
vitórias = wc[wc['vencedor']!='Empate']['vencedor'].value_counts()
vitórias.head(10)
```

```
# 8) Gráfico: distribuição de gols por partida (gols totais)
wc['gols_totais'] = wc['gols_1'] + wc['gols_2']
plt.figure(figsize=(8,4))
wc['gols_totais'].hist(bins=range(0,11))
plt.title('Distribuição de gols por partida')
plt.xlabel('Gols totais')
plt.ylabel('Contagem')
plt.show()
```

```
# 9) Exporte um resumo para CSV (resumo por time)
resumo_time = pd.DataFrame({
    'gols_marcados': long.groupby('time')['gols'].sum(),
    'gols_por_partida': long.groupby('time')['gols'].mean(),
    'vitórias': vitórias
}).fillna(0).sort_values('gols_marcados', ascending=False)
resumo_time.reset_index().head(20)
```
